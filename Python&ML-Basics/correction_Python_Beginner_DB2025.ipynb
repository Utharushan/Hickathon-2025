### <font size=5><u>**If-else statement**</u></font> <br>

Conditional statements are created using the `if` and `else` keywords.

An if-else statement executes code if the `if` statement is True. <br>
It provides an alternate block of code with `else` if the statement is False (not mandatory).

In the following example, the `number > 0` statement is tested.

<img src = https://www.programiz.com/sites/tutorial2program/files/python-if-else.png width = "450" height = "250" >

*Don't forget to indent code inside a conditional statement !*


```python
# IF-ELSE statement to check if Hi! is in the list
hi_list = ["Hi!","PARIS", "Data", "Bootcamp"]

for elem in hi_list:
    if elem == "Hi!":
        print("Yes")
    else:
        print("No")
```

## <font color='#023F7C'>**1.3. For-loops**</font>
**For-loops** allow blocks of code to be repeated a fixed number of times. <br>
They are created using the `for` Python keyword.

*Don't forget to indent the block of code that runs inside a for-loop statement !*



```python
for i in [2, 4, 6, 8]: # for-loop on a list
    print(i)
```

There are functions/methods commonly used with for-loops: <br>
- `range()`: Create a list/range of integers you can iterate on with a loop (function). <br> *Example: Parse through the indexes of an iterable*.

- `.append()`: Add elements at the end of a list with a loop (list method).  <br> *Example: Save the results of a for-loop in a list*.

```python
# Print the indexes of a list
list_numbers = ["a", "b", "c", "d"]

for i in range(len(list_numbers)):
    print(i)
```

```python
# Start with an empty list
list_e = []

# Add the square values of a list of integers to it
for i in [1, 2, 3, 4, 5]:
    list_e.append(i**2)

list_e
```

To create a custom function, you need to add:
- A `def` keyword to define the function,
- **Parameters** that will be the input values of your function,
- A **return statement** to be returned once the block of code has run.

```python
# Create a function with the name 'add_numbers'
# that computes the sum of 2 numbers 'num_1', 'num_2'
# and return this sum

def add_numbers(num_1, num_2):
  return num_1 + num_2
```

```python
# Create a function 'compute_squared' that computes the squared value of
# a number 'num' as input

def compute_squared(num):
  return num**2
```

```python
# Call this function and compute the squared value of 10
compute_squared(10)
```

```python
# Create a function 'square_list' that computes the squared value of
# each element in a list, 'list_input'
# 1. Define an empty list 'result'
# 2. Write a for loop to iterate over the 'list_input'
# 3. Compute the squared value elem**2 + append result using .append()
# 4. Return the result list

def square_list(list_input):
  result = []
  for elem in list_input:
    result.append(elem**2) # save results to an empty list using .append()
  return result # return the result list
```

```python
square_list([2, 4, 6, 8])
```

### <u>**<font size='5'>Indexing a DataFrame**</font> </u><br>

DataFrames are ordered collections of data. <br>
They can be filtered using label and integer-based indexing (`.loc[]`, `.iloc[]` respectively).

#### **`df.loc[conditional_statement]`** <br>
**Select a group of rows in a DataFrame using a conditional statement**. <br>
The conditional statement should apply to columns (and not rows). <br>


```python
# Select every female athletes using "Sex"
dataset.loc[dataset["Sex"]=="F"]
```

Bitwise Operators can also be used to add multiple statements to `.loc[]`.
- `&`: AND
- `|`: OR
- `~`: NOT

You can write several conditional statements as following:

`df.loc[(conditional_statement_1) BITWISE_OPERATOR (conditional_statement_2)]`

```python
# Select every female athlete who is over the age of 30
dataset.loc[(dataset["Sex"]=="F") & (dataset["Age"]>30)]
```

#### **`df.iloc[row_indices, column_indices]`**
<br>

**Filter rows in a DataFrame based on positional index**. <br>

This method only accepts integer indices for rows and columns.

```python
# Select rows between indexes 10 and 20
dataset.iloc[10:21]
```

### **<font size='5'><u>Manipulating Dataframes</u>**</font> <br>

Pandas has a large collection of methods to **manipulate DataFrames**. <br>
These methods include sorting columns, reseting indexes, converting columns to lists, ... <br>

>*Note that Pandas' methods are usually stacked one after the other in a single line of code*.

| Pandas Function | Description | Parameters |
|----------|-------------|------------|
| `.head()`/`.tail()`| Select the first or last *n* rows  | `n`: Number of rows to select, default n=5 |
| `.drop()`| Drop rows or columns from a dataframe | `index`: Indexes to drop <br> `columns`: Columns to drop |
| `.sort_values()`| Order rows based on the values of one or more columns | `by`: Column names to sort by <br> `ascending`: Way of sorting (bool) <br> `inplace`: Perform operation in place (bool)|
| `.value_counts()`| Count the number of rows with each unique variable value | `normalize`: Normalize frequencies (bool) <br> `sort`: Sort by frequencies
| `.reset.index()`| Reset index to row numbers, moving index to columns | `drop`: Keep or drop previous index to df (bool) <br> `inplace`: Apply modifications directly to df (bool)
| `.unique()` | Get the unique values in a DataFrame or Series (column) |
| `.nunique()` | Get the number of unique values in a DataFrame or Series (column) |
| `.to_frame()` | Convert a pandas series to a dataframe |
| `.astype()` | Convert a pandas object (series, dataframe) to a specified dtype | `dtype`: Dtype to cast

```python
# Plot the distribution of ages using sns.kdeplot
sns.kdeplot(data=dataset, x="Age", fill="Age")
plt.title("Distribution of athlete ages")
plt.show()
```

Let's plot the number of medals won by each country (NOC) using seaborn's bar plot. <br>
We will only keep countries who have at least won 50 medals to make the plot readable.

```python
df_nb_medals_country = dataset.groupby("NOC")["has_medal"].sum().reset_index()
df_nb_medals_country.columns = ["Country", "has_medal"]
```

```python
# Keep countries with at least 50 medals
df_nb_medals_country = df_nb_medals_country.loc[df_nb_medals_country["has_medal"]>=50]
df_nb_medals_country
```

```python
# Plot the number of medals won per country (NOC)

plt.figure(figsize=(8, 12))
sns.barplot(data=df_nb_medals_country, x="has_medal", y="Country", hue="has_medal"); # hue: color bars based on the number of medals

# Add matplotlib components (title, ylabel)
plt.title("Number of olympic medals per country")
plt.ylabel("");
plt.legend(title="Number of medals")
plt.show()
```
